```python
# Convert 'TotalCharges' to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce') # here we use errors='coerce' parameter to tell to_numeric method to convert any value like some string value that is not been converted into a number to be converted into NaN value.
print(df.dtypes)
```

```python
# Drop rows with missing values in 'TotalCharges'
df.dropna(subset=['TotalCharges'], inplace=True)

print("After handling TotalCharges missing values:")
print(df.isnull().sum())
```

**2.OneHotEncoding**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Perform one-hot encoding on categorical columns excluding 'customerID' and 'Churn'
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

categorical_cols_for_encoding = [col for col in df.columns if df[col].dtype == 'object' and col not in ['customerID', 'Churn']] # Select categorical columns excluding 'customerID' and 'Churn'

encoded_values = encoder.fit_transform(df[categorical_cols_for_encoding])


new_encoded_cols = encoder.get_feature_names_out(categorical_cols_for_encoding)


encoded_df = pd.DataFrame(encoded_values, columns=new_encoded_cols, index=df.index)

numerical_and_id_cols = df.select_dtypes(include=np.number).columns.tolist() + ['customerID']

df_encoded = pd.concat([df[numerical_and_id_cols], encoded_df], axis=1)

df_encoded['Churn'] = df['Churn']


display(df_encoded.head())
print(df_encoded.shape)
```

```python
# Ensure only the original 'Churn' column (object type) is present before encoding
cols_to_drop = [col for col in df_encoded.columns if 'Churn' in col and col != 'Churn']
df_encoded = df_encoded.drop(columns=cols_to_drop)

# Encode the 'Churn' column and rename it to 'Churn'
df_encoded['Churn'] = df_encoded['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

display(df_encoded.head())
```

```python
df_encoded.info()
```

```python
# Convert the data type of encoded features to int8 to reduce memory usage
for col in encoded_df.columns:
    encoded_df[col] = encoded_df[col].astype('int8')

display(encoded_df.head())
encoded_df.info()
```

**QDA**

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# QDA-To classify data points into categories based on the features
# It assumes each class follows a normal distribution
# Estimates a seperate covariance matrix for each class
# Calculates the probability that a data point belongs to each class
# Assigns the point to the class with the highest probability
```

**when to use QDA**

```python
# When your data shows non-linear separation between classes
# When each class has distinct variance patterns
```

```python
# Step 4: Analyze Model Performance with Other Metrics (continued)
# Train and evaluate QDA after SMOTE and handling multicollinearity
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis



qda=QuadraticDiscriminantAnalysis(reg_param=0.1) # Here the regularization parameter is used to add a small amount of regularization to the covariance matrix calculation

categorical_cols_for_encoding = [col for col in df.columns if df[col].dtype == 'object' and col not in ['customerID', 'Churn']]

encoded_col_names = encoder.get_feature_names_out(categorical_cols_for_encoding)

numerical_col_names = df.select_dtypes(include=np.number).columns.tolist()

all_encoded_cols = numerical_col_names + list(encoded_col_names)


# Drop one column from each set of one-hot encoded features to avoid multicollinearity.
cols_to_drop_from_encoded = [
    'gender_Male', 'Partner_No', 'Dependents_No', 'PhoneService_No',
    'MultipleLines_No phone service', 'InternetService_No',
    'OnlineSecurity_No internet service', 'OnlineBackup_No internet service',
    'DeviceProtection_No internet service', 'TechSupport_No internet service',
    'StreamingTV_No internet service', 'StreamingMovies_No internet service',
    'Contract_Two year', 'PaperlessBilling_No', 'PaymentMethod_Mailed check'
]


X_imputed_df = pd.DataFrame(X_imputed, columns=all_encoded_cols)
X_imputed_dropped = X_imputed_df.drop(columns=cols_to_drop_from_encoded)


X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=[col for col in X_train.columns if col != 'customerID']) # Correct columns for test set
X_test_scaled_dropped = X_test_scaled_df.drop(columns=cols_to_drop_from_encoded)


qda.fit(X_imputed_dropped, y_train_resampled)

print(f"QDA accuracy after SMOTE and handling multicollinearity: {qda.score(X_test_scaled_dropped, y_test)}")
```

```python
# Column transformerss are useful for applying different preprocessing steps on different columns in a ML pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Identify numerical and categorical columns from the original df (excluding customerID and Churn)
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

categorical_cols = [col for col in categorical_cols if col not in ['customerID', 'Churn']]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='drop' # Drop all other columns not specified in transformers, including 'customerID'
)
```

**Hyperparameter tunning**

```python
# The purpose of hyperparameter tuning is to find the optimal settings for a machine learning model that maximize its performance—typically measured by metrics like accuracy, precision, recall, F1-score, or AUC.

param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10]
}

print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
import numpy as np


# Define cv for cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

numerical_cols_X = X.select_dtypes(include=np.number).columns.tolist()
categorical_cols_X = X.select_dtypes(include='object').columns.tolist()

preprocessor_X = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols_X),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols_X)
    ],
    remainder='passthrough'
)

# Redefine the imblearn pipeline with the corrected preprocessor
imb_pipeline = ImbPipeline(steps=[
    ('preprocessing', preprocessor_X),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])


# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=imb_pipeline,
                           param_grid=param_grid,
                           cv=cv,
                           scoring='f1',  # USE f1-score as it's suitable for imbalanced data
                           n_jobs=-1)


grid_search.fit(X_train, y_train)

print("Best hyperparameters found:", grid_search.best_params_)
print("Best cross-validation F1-score:", grid_search.best_score_)
```

**Different Metrics and their evaluation**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy on the test set: {accuracy:.4f}")
print(f"Precision on the test set: {precision:.4f}")
print(f"Recall on the test set: {recall:.4f}")
print(f"F1-score on the test set: {f1:.4f}")
print("\nConfusion Matrix on the test set:")
print(conf_matrix)

# Accuracy-Can be misleading due to imbalanced data in this case
# Precision-Can be prioritize if we want to avoid contacting customers who won't churn
# Recall-We want to find out as many numbers of customers who will churn as possible then we should prioritize it
# F1 score-It is the harmonic mean of precision and recall giving a more balanced metrics and valuable for imbalanced dataset
# From the above 4 Recall and F1 score is important for this scenario
```
