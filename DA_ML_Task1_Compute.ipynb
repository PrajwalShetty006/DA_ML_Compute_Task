```python
# Convert 'TotalCharges' to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce') # here we use errors='coerce' parameter to tell to_numeric method to convert any value like some string value that is not been converted into a number to be converted into NaN value.
print(df.dtypes)
```

```python
# Drop rows with missing values in 'TotalCharges'
df.dropna(subset=['TotalCharges'], inplace=True)

print("After handling TotalCharges missing values:")
print(df.isnull().sum())
```

**2.OneHotEncoding**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Perform one-hot encoding on categorical columns excluding 'customerID'
encoder=OneHotEncoder(handle_unknown='ignore',sparse_output=False)

categorical_cols = [col for col in df.columns if df[col].dtype == 'object' and col != 'customerID'] # Select categorical columns excluding 'customerID' from the original df

# Fit and transform the categorical columns using OneHotEncoder
encoded_values=encoder.fit_transform(df[categorical_cols])

# Get the new column names
new_cols=encoder.get_feature_names_out(categorical_cols)

encoded_df = pd.DataFrame(encoded_values, columns=new_cols, index=df.index) # Create a new DataFrame with the encoded values and new column names

df_dropped_categorical = df.drop(columns=categorical_cols) # Drop the original categorical columns from df

df_encoded = pd.concat([df_dropped_categorical, encoded_df], axis=1) # Concatenate the original numerical columns and the new encoded columns

display(df_encoded.head())
print(df_encoded.shape)
```

```python
# Display information about the encoded DataFrame
df_encoded.info()
```

```python
# Convert the data type of encoded features to int8 to reduce memory usage
for col in encoded_df.columns:
    encoded_df[col] = encoded_df[col].astype('int8')

display(encoded_df.head())
encoded_df.info()
```

```python
# LDA-It focuses on maximizing the seperability among known categories.
# LDA creates a new axis and projects the data onto the new axis in a way to maximize the seperation of the two categories.
# How does LDA creates new axis,it is based on the two criteria.
# 1.Maximize the distance between means of the X and Y axis points in case of 2-D to 1-D
# 2.Minimize the variation within each category.
```

**QDA**

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# QDA-To classify data points into categories based on the features
# It assumes each class follows a normal distribution
# Estimates a seperate covariance matrix for each class
# Calculates the probability that a data point belongs to each class
# Assigns the point to the class with the highest probability
```

**when to use QDA**

```python
# When your data shows non-linear separation between classes
# When each class has distinct variance patterns
```
